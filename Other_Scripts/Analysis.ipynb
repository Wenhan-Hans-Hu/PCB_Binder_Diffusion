```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data
file_path = 'rmsd_results.csv'
df = pd.read_csv(file_path)

print("DataFrame shape:", df.shape)
print("Available columns:", df.columns.tolist())

# Create subplots with custom width ratios (e.g., 3:1 for histogram:boxplot)
fig, axes = plt.subplots(1, 2, figsize=(12, 6), gridspec_kw={'width_ratios': [3, 1]})

# Histogram - give it more space
axes[0].hist(df['RMSD'].dropna(), bins=50, edgecolor='black')
axes[0].set_title('Ligand Only All-Atom RMSD')
axes[0].set_xlabel('RMSD')
axes[0].set_ylabel('Frequency')
axes[0].grid(True)

# Boxplot - now narrower
axes[1].boxplot(df['RMSD'].dropna(), vert=True, patch_artist=True,
                boxprops=dict(facecolor='lightblue', color='black'),
                medianprops=dict(color='red'),
                whiskerprops=dict(color='black'),
                capprops=dict(color='black'),
                flierprops=dict(marker='o', markerfacecolor='orange', markersize=5, linestyle='none'))

axes[1].set_title('Box Plot of RMSD per AAR Pair')
axes[1].set_ylabel('RMSD')
axes[1].grid(True, axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Load data
file_path = 'rmsd_results.csv'
df = pd.read_csv(file_path)

print("DataFrame shape:", df.shape)
print("Available columns:", df.columns.tolist())

# Set up the figure with custom width ratio
fig, axes = plt.subplots(1, 2, figsize=(12, 6), gridspec_kw={'width_ratios': [3, 1]})

# Histogram (left)
axes[0].hist(df['RMSD'].dropna(), bins=50, edgecolor='black')
axes[0].set_title('Ligand Only All-Atom RMSD')
axes[0].set_xlabel('RMSD')
axes[0].set_ylabel('Frequency')
axes[0].grid(True)

# Box + Strip Plot (right)
sns.boxplot(y=df['RMSD'], ax=axes[1], width=0.3, color='lightblue',
            boxprops=dict(alpha=0.6), fliersize=0)

# Overlay individual data points with jitter
sns.stripplot(y=df['RMSD'], ax=axes[1], color='orange', size=4, alpha=0.6, jitter=0.2)

axes[1].set_title('Box Plot of RMSD per AAR Pair')
axes[1].set_ylabel('RMSD')
axes[1].set_xticks([])  # Remove x-axis ticks since it's a single category
axes[1].grid(True, axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()

mean_rmsd = df['RMSD'].mean()
print("Average RMSD:", mean_rmsd)
```

```python
from rdkit import Chem
from rdkit.Chem import AllChem

# Load molecules from PDB files
mol_tar = Chem.MolFromPDBFile("tar.pdb", removeHs=True)
mol_ref = Chem.MolFromPDBFile("ref.pdb", removeHs=False)

# Generate 3D conformers if not already present (optional)
# You can skip this if your PDBs already contain good coordinates
# AllChem.EmbedMolecule(mol_tar)
# AllChem.EmbedMolecule(mol_ref)

# Check if molecules loaded properly
if mol_tar is None or mol_ref is None:
    raise ValueError("Failed to load one or both PDB files.")

# Make sure atoms are in the same order for correct alignment
# If they are not, you’ll need to map them (see advanced section below)

# Align tar to ref and compute RMSD
rmsd = AllChem.GetBestRMS(mol_tar, mol_ref)

print(f"RMSD between tar.pdb and ref.pdb: {rmsd:.3f} Å")
```
